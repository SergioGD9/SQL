### 6 Empleados

### 8 Compañias

**8.1 Creamos una vista llamada  Customer_with_companies, donde estarán incluidos todos los resultados salvo cuando Company sea NULL**

```python
cursor.execute("CREATE VIEW IF NOT EXISTS Customer_with_companies AS \
                SELECT * FROM Customer WHERE Company IS NOT NULL")
```

**8.2 Ordenamos los resultados por orden alfabetico**

```python
cursor.execute("SELECT * FROM Customer_with_companies ORDER BY Company")
orden_alfabetico = cursor.fetchall()
print("Resultados ordenados por orden alfabético de Company:\n")
for row in orden_alfabetico:
    print(row)
```

**8.3 Compañias de Brasil**

```python
cursor.execute("SELECT DISTINCT Company FROM Customer_with_companies WHERE Country = 'Brazil'")
Brasil = cursor.fetchall()
print("Compañías de Brasil:\n")
for row in Brasil:
    print(row[0])
```

### 10 Buscando Facturas

**10.1 Muestra: InvoiceId, nombre del cliente y BillingCountry**

```python
cursor.execute("SELECT InvoiceId, Customer.FirstName || ' ' || Customer.LastName AS CustomerName, BillingCountry \
                FROM Invoice \
                INNER JOIN Customer ON Invoice.CustomerId = Customer.CustomerId \
                ORDER BY Total DESC")
resultado = cursor.fetchall()
print("InvoiceId, nombre del cliente y BillingCountry ordenados de mayor a menor por Total:\n")
for row in resultado:
    print(row)
```

**10.2 Pais que más a facturado**

```python
cursor.execute("SELECT BillingCountry, SUM(Total) AS TotalSales \
                FROM Invoice \
                GROUP BY BillingCountry \
                ORDER BY TotalSales DESC \
                LIMIT 1")
facturado = cursor.fetchone()
print("País que más ha facturado:\n")
print(facturado[0])
```
